## Removing all entries that are not either White or Black

```python
df = df[
(df["African_American"] == 1) | ((df["African_American"] == 0) & (df["Asian"] == 0) & (df["Hispanic"] == 0) & (df["Native_American"] == 0) & (df["Other"] == 0))
]
```

## Splitting into two datasets, one black and one white

```python
#since we removed all races except the two, we can assume all non Black entries are White

df_Black = df[df["African_American"] == 1]
df_White = df[df["African_American"] == 0]
```

## Now finding false positives (FPR) via the following formula:
$$FPR = \frac{FP}{FP + TN}$$

Where FP is the number of false positives (predicted by algorithm to re-offend, and did not re-offend), and TN is true negatives (predicted by algorithm to not re-offend, and did not re-offend).


```python
FP_Black = len(df_Black[(df_Black["Two_yr_Recidivism"] == 0) & (df_Black["score_factor"] == 1)])
TN_Black = len(df_Black[(df_Black["Two_yr_Recidivism"] == 0) & (df_Black["score_factor"] == 0)])

FPR_Black = FP_Black/(FP_Black + TN_Black)
FPR_Black
```

```python
FP_White = len(df_White[(df_White["Two_yr_Recidivism"] == 0) & (df_White["score_factor"] == 1)])
TN_White = len(df_White[(df_White["Two_yr_Recidivism"] == 0) & (df_White["score_factor"] == 0)])

FPR_White = FP_White/(FP_White + TN_White)
FPR_White
```
